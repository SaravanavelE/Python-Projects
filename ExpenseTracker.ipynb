```python
def display_menu():
    print("\nExpense Tracker:")
    print("1. Add Expense")
    print("2. Edit Expense")
    print("3. Delete Expense")
    print("4. View All Expenses")
    print("5. Exit")

def add_expense(expenses):
    expense = input("Enter expense name: ").strip()
    if expense in expenses:
        print("Expense is already exists.")
    else:
        amount = float(input("Enter expense amount: "))
        expenses[expense] = amount
        print(f"Expense {expense} added successfully.")
    print("----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")

def edit_expense(expenses):
    expense = input("Enter the expense name to edit: ").strip()
    if expense in expenses:
        new_amount = float(input("Enter the new amount: "))
        expenses[expense] = new_amount
        print(f"Expense {expense} updated successfully.")
    else:
        print("Expense not found.")
    print

def delete_expense(expenses):
    expense = input("Enter the expense name to delete: ").strip()
    if expense in expenses:
        del expenses[expense]
        print(f"Expense {expense} deleted successfully.")
    else:
        print("Expense not found.")
    print("----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")

def view_all_expenses(expenses):
    if not expenses:
        print("No expenses available.")
    else:
        print("\nAll Expenses:")
        for expense, amount in expenses.items():
            print(f"{expense}: {amount}")
        print("Total amount :", sum(expenses.values()))
    print("----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")

def main():
  expenses = {}
  while True:
    display_menu()
    choice = input("Enter your choice: ").strip()
    if choice == "1":
      add_expense(expenses)
    elif choice == "2":
      edit_expense(expenses)
    elif choice == "3":
      delete_expense(expenses)
    elif choice == "4":
      view_all_expenses(expenses)
    elif choice == "5":
      print("Exiting Expense Tracker.")
      break
    else:
      print("Invalid choice. Please enter a valid option.")

main()
```
